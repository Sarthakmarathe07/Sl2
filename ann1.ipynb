```python
import numpy as np
import matplotlib.pyplot as plt

# Activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.maximum(alpha * x, x)

def softmax(x):
    exp_x = np.exp(x - np.max(x, axis=-1, keepdims=True))
    return exp_x / np.sum(exp_x, axis=-1, keepdims=True)

def tanh(x):
    return np.tanh(x)

```

```python
x = np.linspace(-5, 5, 100)

# Compute activation values
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_leaky_relu = leaky_relu(x)
y_tanh = tanh(x)
```

```python
def plot_activation(x, y, title, ax):
    ax.plot(x, y, label=title, linewidth=2)
    ax.axhline(0, color='black', linewidth=0.5, linestyle='dashed')
    ax.axvline(0, color='black', linewidth=0.5, linestyle='dashed')
    ax.set_title(title)
    ax.set_xlabel('Input')
    ax.set_ylabel('Output')
    ax.legend()
    ax.grid(True)
```

```python
plt.figure(figsize=(12, 8))

# Sigmoid
ax1 = plt.subplot(2, 2, 1)
plot_activation(x, y_sigmoid, 'Sigmoid', ax1)

# ReLU
ax2 = plt.subplot(2, 2, 2)
plot_activation(x, y_relu, 'ReLU', ax2)

# Leaky ReLU
ax3 = plt.subplot(2, 2, 3)
plot_activation(x, y_leaky_relu, 'Leaky ReLU (α=0.01)', ax3)

# Tanh
ax4 = plt.subplot(2, 2, 4)
plot_activation(x, y_tanh, 'Hyperbolic Tangent (Tanh)', ax4)

plt.tight_layout()
plt.show()
```

```python
# Softmax requires a different approach since it's multi-dimensional
x_softmax = np.array([[i] for i in range(-5, 6)]) # 1D input for softmax
y_softmax = softmax(x_softmax.T).T # Compute softmax
plt.subplot(2, 2, 4)
plt.bar(range(-5, 6), y_softmax.flatten(), label='Softmax')
plt.title('Softmax')
plt.xlabel('Input')
plt.ylabel('Probability')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```
