```python
import numpy as np
import matplotlib.pyplot as plt

# Activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def tanh(x):
    return np.tanh(x)


def swish(x):
    return x * sigmoid(x)



# Softmax function
def softmax(x):
    exp_x = np.exp(x - np.max(x))  # subtract max for numerical stability
    return exp_x / np.sum(exp_x)

# Input range
x = np.linspace(-5, 5, 100)


# Compute activations
activations = {
    'Sigmoid': sigmoid(x),
    'ReLU': relu(x),
    'Leaky ReLU (=0.01)': leaky_relu(x),
    'Tanh': tanh(x)
}

# Plotting
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

for ax, (name, y) in zip(axes.flatten(), activations.items()):
    ax.plot(x, y, label=name, color='blue')
    ax.axhline(0, color='black', linestyle='--', linewidth=0.5)
    ax.axvline(0, color='black', linestyle='--', linewidth=0.5)
    ax.set_title(name)
    ax.legend()
    ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
x_softmax = np.linspace(-2, 2, 10)  # 10 points only
y_softmax = softmax(x_softmax)

# Plot Softmax
plt.figure(figsize=(6, 4))
plt.stem(x_softmax, y_softmax, linefmt='blue', markerfmt='bo', basefmt='k')  # stem plot looks better
plt.axhline(0, color='black', linestyle='--', linewidth=0.5)
plt.axvline(0, color='black', linestyle='--', linewidth=0.5)
plt.title('Softmax')
plt.xlabel('Input')
plt.ylabel('Probability')
plt.legend(['Softmax'])
plt.grid(True)
plt.show()
```
